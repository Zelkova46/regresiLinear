```python
import numpy as np
import math 
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
dataset = pd.read_csv('merkhp2.csv')
x = dataset.iloc[:, 1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

```python
print(y)
```

```python
x_train = []
i = 0
for i in range(len(x)): 
  if i != 3 and i != 5:
    x_train.append(x[i])

  i = i + 1

print(x_train)
```

```python
y_train = []
i = 0
for i in range(len(y)): 
  if i != 3 and i != 5:
    y_train.append(y[i])

  i = i + 1

print(y_train)
```

```python
x_test = []
x_test.append(x[3])
x_test.append(x[5])
```

```python
x_test
```

```python
y_test = []
y_test.append(y[3])
y_test.append(y[5])
```

```python
y_test
```

### mencari nilai slope, gradian, m


```python
XY = []
for i in range(len(x_train)):
  perkalianXY = x_train[i] * y_train[i]

  XY.append(perkalianXY)
  i = i + 1
```

```python
print(XY)
```

```python
sigmaXY = sum(XY)
print(sigmaXY)
```

```python
XX = []
for i in range(len(x_train)):
  Kuadrat = x_train[i] * x_train[i]
  XX.append(Kuadrat)
  i = i + 1
```

```python
print(XX)
```

```python
sigmaXX = sum(XX)
print(sigmaXX)
```

```python
n = len(x_train)
```

```python
sigmaX = sum(x_train)
print(sigmaX)
```

```python
sigmaY = sum(y_train)
print(sigmaY)
```

```python
m_pembilang = (n * sigmaXY) - (sigmaX * sigmaY)
```

```python
m_penyebut = (n * sigmaXX) - (sigmaX * sigmaX)
```

```python
print(m_penyebut)
```

```python
print(m_pembilang)
```

```python
m = m_pembilang / m_penyebut
print(m)
```

### mencari nilai irisan, konstanta, C


```python
c_pembilang = sigmaY - (m * sigmaX)
print(c_pembilang)
```

```python
c_penyebut = n
```

```python
print(n)
```

```python
c = c_pembilang / c_penyebut
print(c)
```

### oke m udah c udah bisa dibuat persamaan linear


```python
x = np.linspace(0,12, 20)
y = m * x + c
plt.plot(x, y, '-r', label="y = {m}x + {c}".format(m = m, c = c))
plt.plot(x_train, y_train, 'ro')

plt.title("Graph of y = {m}x + {c}".format(m = m, c = c))
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')

plt.show()
```
