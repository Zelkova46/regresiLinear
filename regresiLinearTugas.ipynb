### konversi kolom menjadi list

```python
x = dataku['Kamera'].values.tolist()
print(x)  # x variable untuk megapixel kamera
```

```python
y = dataku['Harga'].values.tolist()
print(y) # y variable untuk harga kamera
```

### mencari mean x dan mean y

```python
rerata_x = sum(x) / len(x)
```

```python
rerata_y = sum(y) / len(y)
```

### mencari nilai w aksen dengan rumus

### menghitung bagian pembilang rumus mencari w


```python
n = len(x)
j_atas = 0
for i in range(n):
  j_atas = j_atas + (x[i] * (y[i] - rerata_y))

j_atas
```

### menghitung bagian penyebut rumus mencari w


```python
m = len(y)
j_bawah = 0
for i in range(m):
  j_bawah = j_bawah + (x[i] * (x[i] - rerata_x))

j_bawah
```

```python
w = j_atas / j_bawah
print(w)
```

wakhir.png

### cara 2 menggunakan fitur lain dari looping for

```python
bag_atas = sum([kam * (har - rerata_y) for (kam, har) in zip(x,y)])
bag_bawah = sum([kam * (kam - rerata_x) for (kam, har) in zip(x,y)])
w = bag_atas / bag_bawah
print(w)
```

### Hitung nilai a dengan rumus

```python
a = rerata_y - (w *rerata_x)
print(a)
```

### Hitung prediksi harga H4 dan H6 menurut Machine Learning

```python
kameraH4H6 = [9 , 5]
harga_H4 = a + (w * kameraH4H6[0])
harga_H6 = a + (w * kameraH4H6[1])
```

```python
print("Harga HP4 dengan kamera " , kameraH4H6[0], " adalah " , harga_H4)
print("Harga HP6 dengan kamera " , kameraH4H6[1], " adalah " , harga_H6)
```

### cari harga HP menurut pemodelan machine learning yang dilakukan

```python
val_y= []
for i in range(m):
  yi = a + (w * x[i])
  val_y.append(yi)
  
```

```python
print(val_y)
```
